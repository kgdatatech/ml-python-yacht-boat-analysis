```python
#  Create/save project folder path
path = r"Boat Sales Analysis"

# Read data (csv file)
df = pd.read_csv(os.path.join(path, '02 Data', 'Prepared Data', 'boat_sales_clusters.csv'), index_col=False)

# Read data (csv file)
df1 = pd.read_csv(os.path.join(path, '02 Data', 'Prepared Data', 'boat_sales_final1.csv'), index_col=False)
```

```python
df = df.drop(columns = ['Unnamed: 0'])

df.head(5)
```

```python
# Rename column
df.rename(columns = {'Year Built' : 'YearEst'}, inplace = True)
df.rename(columns = {'# Views' : '#Views'}, inplace = True)
df.rename(columns = {'clusters' : 'ClusterID'}, inplace = True)
df.rename(columns = {'cluster' : 'ClusterGrp'}, inplace = True)

df.head(5)
```

```python
df1 = df1.drop(columns = ['Unnamed: 0'])

df1.head(5)
```

```python
# Rename column
df1.rename(columns = {'Specifications' : 'Cond'}, inplace = True)
df1.rename(columns = {'Year Built' : 'YearEst'}, inplace = True)
df1.rename(columns = {'Year Built Condition' : 'AgeCond'}, inplace = True)
df1.rename(columns = {'Length Condition' : 'LenCond'}, inplace = True)
df1.rename(columns = {'Width Condition' : 'WidCond'}, inplace = True)
df1.rename(columns = {'Price Condition' : 'ValCond'}, inplace = True)
df1.rename(columns = {'Views Condition' : 'VisCond'}, inplace = True)
df1.rename(columns = {'State/Region' : 'StRegion'}, inplace = True)
df1.rename(columns = {'# Views' : '#Views'}, inplace = True)

df1.head(5)
```
